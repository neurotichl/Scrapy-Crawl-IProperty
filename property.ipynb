```python
import pandas as pd
import folium
import re
```

```python
df1 = pd.read_csv('house_cheap.csv')
df2 = pd.read_csv('house_exp.csv')
df = df1.append(df2, ignore_index=True)
```

```python
def facilities(x):
    try:
        return x.replace('Bedroom(s)','Bed').replace('Bathroom(s)','Bath').replace('Parking Bay(s)','Park')
    except:
        pass
    
df = df[df['price'].map(lambda x: len(x)>8)]
df['facilities'] = df['amenities'].map(facilities)
df = df[df['lat'].notnull()]
```

```python
df.reset_index(drop=True).reset_index()
```

```python
def add_frame(count,name,price,size,amenities,link,**kwargs):
    html="""
        <pre>
{0}
Name      : {1}
Price     : {2}
Amenities : {3}
sq. ft    : {4}
link      : {5}
        </pre>
        """
    return folium.IFrame(html= html.format(count,name,price,amenities,size,link), width=400, height=100)

color_group = {'High':'red','Low':'blue'}
m = folium.Map(location=[3.139, 101.6869], zoom_start = 10)

count=1
for row in df.to_dict(orient='records'):
    iframe = add_frame(count,**row)
    popup = folium.Popup(iframe, max_width=2650)
    count+=1
    folium.Marker([row['lat'], row['lon']], popup=popup, icon=folium.Icon(color=color_group[row['prize_range']])).add_to(m)
```
